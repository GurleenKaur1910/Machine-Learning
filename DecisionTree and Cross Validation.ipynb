```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##Boston House Pricing Dataset
from sklearn.datasets import load_diabetes
diabetes_df=load_diabetes()
```

```python
diabetes_df
```

```python
#independent features and dependent features
X=pd.DataFrame(diabetes_df.data,columns=diabetes_df.feature_names)
y=diabetes_df.target
```

```python
X,y
```

```python
X.head()
```

```python
##train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_pred,y_test)
```

```python
score
```

```python
#Hyperparameter tuning
parameter={
    'criterion':['squared_error','friedman_mse','absoute_error','poisson'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,6,7,8,10,11,12],
    'max_features':['auto','sqrt','log2']
}
regressor=DecisionTreeRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
regressorcv=GridSearchCV(regressor,param_grid=parameter,cv=5,scoring='neg_mean_squared_error')
```

```python
regressorcv.fit(X_train,y_train)
```

```python
regressorcv.best_params_
```

```python
y_pred=regressorcv.predict(X_test)
```

```python
r2_score(y_pred,y_test)
```
